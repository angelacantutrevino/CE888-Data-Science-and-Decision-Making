# **Recommend Movie Function**

```python
def recommend_movie(user, num_recommendations):
  user = np.array(df_users[df_users['User'] == user].loc[:,'Factor1':])
  score = pd.DataFrame(columns=['ratings','Movie'])
  movie_id = df_items['Movie ID']
  j = 0
  for i in df_items.drop(['Movie ID'],axis=1).values:
    rating = user.dot(i[1:].reshape(-1,1))
    movies = {'ratings':rating,'Movie':i[0],'Movie ID':movie_id[j]}
    score = score.append(movies, ignore_index=True)
    j += 1
  return score.sort_values('ratings', ascending=False).head(num_recommendations)
```

In the movies dataset, if you had to recommend 3 movies to the user with ID=525, which ones would you choose?

```python
print(recommend_movie(525,3))
```
